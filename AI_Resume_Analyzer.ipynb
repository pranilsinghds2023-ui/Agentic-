```python
import os
from getpass import getpass

os.environ["GROQ_API_KEY"] = getpass("Enter your Groq API Key: ")
```

```python
from groq import Groq
from pypdf import PdfReader
from google.colab import files

client = Groq(api_key=os.environ["GROQ_API_KEY"])

print("✅ Everything Loaded Successfully")
```

```python
uploaded = files.upload()
```

```python
pdf_file = list(uploaded.keys())[0]

reader = PdfReader(pdf_file)

resume_text = ""

for page in reader.pages:
    text = page.extract_text()
    if text:
        resume_text += text + "\n"

print("Resume Loaded Successfully!")
print("-" * 50)
print(resume_text[:1000])
```

```python
def analyze_resume(resume_text):

    prompt = f"""
You are an expert HR Recruiter and Career Coach.

Analyze the following resume.

Provide the output in this format:

1. Professional Summary

2. Skills Detected

3. Resume Score (out of 100)

4. Missing Skills

5. Improvement Suggestions

6. Five Interview Questions

Resume:

{resume_text}
"""

    response = client.chat.completions.create(
        model="llama-3.3-70b-versatile",
        messages=[
            {
                "role": "user",
                "content": prompt
            }
        ],
        temperature=0.3,
        max_tokens=1500
    )

    return response.choices[0].message.content
```

```python
result = analyze_resume(resume_text)

print(result)
```

```python
with open("Resume_Analysis_Report.txt", "w", encoding="utf-8") as f:
    f.write(result)

print("✅ Report saved as Resume_Analysis_Report.txt")
```
